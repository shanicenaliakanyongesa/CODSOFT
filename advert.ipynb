```python
import pandas
data=pandas.read_csv('Advertising.csv')
data
```

```python
data.isnull().sum()
```

```python
array = data.values
x = array[:, 0:5] 
```

```python
import sklearn
from sklearn.cluster import KMeans
model = KMeans(n_clusters=25, random_state=42)
model.fit(x)
```

```python
means = model.cluster_centers_
clusters = pandas.DataFrame(means, columns = ['No', 'TV','Radio', 'Newspaper','Sales'])
clusters
```

```python
data['no']  = model.labels_   # get every record group number
cluster22 =  data[data['no'] == 22]
cluster22
```

```python
data[data['no'] == 22].to_csv('Cluster22.csv')
```
